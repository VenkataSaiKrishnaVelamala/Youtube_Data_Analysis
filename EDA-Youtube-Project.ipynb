## Creating the Tables like Channel, Videos, Fact and Period

```python
mycursor.execute('CREATE TABLE IF NOT EXISTS CHANNEL (CHANNEL_ID VARCHAR(30) PRIMARY KEY,CHANNEL_NAME VARCHAR(50),SUBSCRIBER_COUNT INT,VIEW_COUNT INT,VIDEO_COUNT INT)')
```

```python
mycursor.execute("CREATE TABLE IF NOT EXISTS VIDEOS (Video_Id VARCHAR(30) PRIMARY KEY,Title VARCHAR(200),PUBLISHED_AT DATE)")
```

```python
mycursor.execute("CREATE TABLE IF NOT EXISTS FACT (,Video_Id VARCHAR(30) ,CHANNEL_ID VARCHAR(30),View_Count INT,Like_Count INT,Comment_Count INT,FOREIGN KEY (CHANNEL_ID) REFERENCES CHANNEL(CHANNEL_ID),FOREIGN KEY (Video_Id) REFERENCES VIDEOS(Video_Id))")
```

```python
mycursor.execute("CREATE TABLE IF NOT EXISTS FACT (ID INT AUTO_INCREMENT PRIMARY KEY,Video_Id VARCHAR(30) ,CHANNEL_ID VARCHAR(30),View_Count INT,Like_Count INT,Comment_Count INT, Period varchar(10),FOREIGN KEY (CHANNEL_ID) REFERENCES CHANNEL(CHANNEL_ID),FOREIGN KEY (Video_Id) REFERENCES VIDEOS(Video_Id))")
```
